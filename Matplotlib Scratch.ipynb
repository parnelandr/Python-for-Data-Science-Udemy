# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

### functional method for creating matplotlib plots

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### object oriented method

```python
fig = plt.figure() #create canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('xlabel')
axes.set_ylabel('ylabel')
axes.set_title('title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #x coord on canvas, y coord on canvas, x width on canvas, y height on canvas

axes1.plot(x,y)
axes1.set_xlabel('xlabel')
axes1.set_ylabel('ylabel')
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

## creating subplots

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

### note on plots

figure is the canvas that the plot goes on

axes are each chart within the canvas

# Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## Lables

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Mytitle')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc = (0.5, 0.3)) # add loc=an integer 0-10 0=matplotlib best 1-10 is top left to bottom right like buttons on a phone 
            #also can pass a tuple with coordinates ie loc=(x,y) x and y are within your compared against your axes parameters
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='red') #common colour names are accepted
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00') #hex codes are accepted
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth=0.3)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth=4, alpha=0.25) #alpha is opacity
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth=4, linestyle='--') #look up options linestyle can be ls and linewidth lw
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=0.5,marker='o',markersize=5,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='blue') 

#look up options for markers, markersize, markerfacecolor, markeredgewidth etc 
#in the Matplotlib Concepts Lecture there's a ton of examples shown 
```

## Axes control and appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
